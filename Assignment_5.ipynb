```python
def calculate_discount(price, customer_type):
    if customer_type.lower() == "regular":
        discount = 0.05
    elif customer_type.lower() == "premium":
        discount = 0.15
    elif customer_type.lower() == "employee":
        discount = 0.25
    else:
        return "Invalid customer type"

    final_price = price - (price * discount)
    return final_price
```
