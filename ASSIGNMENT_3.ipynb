```python
# Attendance Summary Generator

n = int(input("Enter number of interns: "))

present_count = 0
absent_count = 0

for i in range(1, n + 1):
    status = input(f"Enter attendance for Intern {i} (P/A): ").upper()

    if status == "P":
        present_count += 1
    elif status == "A":
        absent_count += 1
    else:
        print("Invalid input, enter P or A")

print("\n--- Attendance Summary ---")
print("Total Present:", present_count)
print("Total Absent:", absent_count)
```
