```python
def calculate_salary(basic_salary, present_days):
    hra = 0
    da = 0
    pf = basic_salary * 0.12

    if present_days >= 3:
        hra = basic_salary * 0.20
        da = basic_salary * 0.10

    net_salary = basic_salary + hra + da - pf
    return net_salary
```
