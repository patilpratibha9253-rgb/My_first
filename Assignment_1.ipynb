```python
def check_intern_eligibility(age, percentage):
    if age >= 18 and percentage >= 60:
        return "Eligible: Age and percentage criteria satisfied."
    elif age < 18:
        return "Not Eligible: Age must be 18 or above."
    elif percentage < 60:
        return "Not Eligible: Graduation percentage must be 60% or above."


# Example function calls
result1 = check_intern_eligibility(20, 75)
print(result1)

result2 = check_intern_eligibility(17, 70)
print(result2)

result3 = check_intern_eligibility(22, 55)
print(result3)
```
