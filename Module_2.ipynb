```python
def attendance_management():
    present_days = 0

    print("Enter attendance for 5 working days (P/A):")
    for day in range(1, 6):
        status = input(f"Day {day}: ").upper()
        if status == 'P':
            present_days += 1

    if present_days >= 4:
        attendance_status = "Regular"
    elif present_days == 3:
        attendance_status = "Warning"
    else:
        attendance_status = "Irregular"

    print(f"Total Present Days: {present_days}")
    print(f"Attendance Status: {attendance_status}\n")

    return present_days
```
