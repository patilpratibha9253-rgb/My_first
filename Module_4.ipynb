```python
def performance_evaluation():
    total_rating = 0

    print("Enter 3 performance ratings (1 to 5):")
    for i in range(1, 4):
        rating = int(input(f"Rating {i}: "))
        total_rating += rating

    average = total_rating / 3

    if average >= 4:
        result = "Excellent"
    elif average >= 3:
        result = "Good"
    else:
        result = "Needs Improvement"

    print(f"Average Rating: {average:.2f}")
    print(f"Performance: {result}\n")
```
