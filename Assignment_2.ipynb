```python
# Input distance
distance = int(input("Enter delivery distance (in km): "))

# Calculate delivery charge
if distance <= 5:
    charge = 0
elif distance <= 15:
    charge = distance * 10
else:
    charge = distance * 20

# Display result
print("Total Delivery Charge: ₹", charge)
```
